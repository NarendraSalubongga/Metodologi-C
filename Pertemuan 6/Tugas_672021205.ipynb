```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

datapenjualan_makananpenutup = ('Donat', 'Pastel', 'Kue Coklat', 'Ice Cream', 'Puding Vanila', 'Brownies', 'Puding Strawberi',
                                'Puding Coklat', 'Ice Cream Nutela', 'Kue Coklat-Keju', 'Kue Wajik', 'Kue Sus', 'Mochi')
jumlah_terjual = (14, 5, 12, 19, 6, 8, 12, 9, 10, 17, 2, 9, 13)

df = pd.DataFrame({
    'Data Penjualan Makanan Penutup': datapenjualan_makananpenutup,
    'Jumlah Terjual': jumlah_terjual,
})

df.sort_values(by='Jumlah Terjual', inplace=True)
x_coords = np.arange(len(df))
colors = ['#0000FF' for _ in range(len(df))]

colors[-13] = '#FF0000'
colors[-12] = '#FF0000'
colors[-11] = '#FF0000'
plt.figure(figsize=(20, 10))

plt.bar(x_coords, df['Jumlah Terjual'], tick_label=df['Data Penjualan Makanan Penutup'], color=colors)
plt.xticks(rotation=90)
plt.ylabel('Jumlah Makanan Penutup Terjual')
plt.title('Jumlah Penjualan Setiap Jenis Makanan Penutup')
plt.show()
```
